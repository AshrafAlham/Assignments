```python
import pandas as pd
import numpy as np
```

```python
# 1- Write a Pandas program to create and display a one-dimensional array-like object containing an array of data using Pandas module
series = pd.Series([1, 3, 5, 7, 9])
series
```

```python
# 2- Write a Pandas program to convert a Panda module Series to Python list and it's type.
print(series.tolist())
print(type(series.tolist()))
```

```python
# 3- Write a Pandas program to compare the elements of the two Pandas Series
series1 = pd.Series([1, 3, 4, 5, 6])
series2 = pd.Series([2, 3, 5, 5, 6])
print(series1 == series2)
```

```python
# 4- Write a Pandas program to change the data type of given a column or a Series.
series3 = pd.Series([1, "python", 4, "data", 6])
print(pd.to_numeric(series3, errors="coerce"))
```

```python
# 5- Write a Pandas program to get the powers of an array values element-wise.
x= np.array([8,5,6,5,6])
y = np.array([8,9,8,3,8])
x_series = pd.Series(x)
y_series = pd.Series(y)
powered = x_series.pow(y_series)
powered
```
