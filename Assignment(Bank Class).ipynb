```
class Bank:
  def __init__(self, accounts, balances):
    self.accounts = accounts
    self.balances = balances

  def register_user(self, username, password):
    if username in self.accounts.keys():
      print(f"this username {username} already exists")
    else:
      self.accounts.update({username: password})
      self.balances.update({username: 0})

  def login_user(self, username, password):
    if username in self.accounts.keys():
      if self.accounts[username] == password:
        print("Login Successful")
        return username
    print("Login Failed")
    return False

  def check_balance(self, username):
    if username in self.accounts.keys():
      print(f"Hey,{username} your balance is: {self.balances[username]}")
    else:
      print("User not found")

  def deposit(self, username, amount):
    if username in self.balances.keys():
      if amount > 0:
        self.balances[username] += amount
      else:
        print("Invalid amount")
    else:
      print("User not found")

  def withdraw(self, username, amount):
    if username in self.balances.keys():
      if 0 < amount < self.balances[username]:
        self.balances[username] -= amount
        return "Withdrawl Successful"
      else:
        print("Invalid amount")
    else:
      print("User not found")


```

```

bank = Bank({}, {})
while True:
  main_menu = input("what do you want: \n1-Login\n2-Register\n3-Exit\n(write 1 or 2 or 3)\n")
  if main_menu == "1":
    user_menu = input("what do you want: \n1-Check Balance\n2-Deposit\n3-Withdraw\n4-Exit\n(write 1 or 2 or 3 or 4)\n")
    if user_menu == "1":
      user_name = input("what is your name? ")
      bank.check_balance(user_name)
    elif user_menu == "2":
      user_name = input("what is your name? ")
      amount = float(input("Hou much money do you want to add to your balance? "))
      bank.deposit(user_name, amount)
    elif user_menu == "3":
      user_name = input("what is your name? ")
      amount = float(input("Hou much money do you want to get from your balance? "))
      bank.withdraw(user_name, amount)
    else:
      break
  elif main_menu == "2":
    user_name = input("what is your name? ")
    user_pass = input("whta is the password? ")
    bank.register_user(user_name, user_pass)
  else:
    break
```
